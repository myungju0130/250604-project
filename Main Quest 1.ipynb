```python
import random

class Account:
    # Q2. 클래스 변수: 계좌 개수
    account_count = 0

    def __init__(self, owner, balance):
        # Q1. 생성자: 예금주, 초기 잔액, 은행이름(SC은행), 랜덤 계좌번호
        self.owner = owner
        self.balance = balance
        self.bank = "SC은행"
        self.account_number = self._generate_account_number()
        Account.account_count += 1  # 클래스 변수로 계좌 개수 카운트

        # Q7. 이자 관련 & Q10. 입출금 내역
        self.deposit_count = 0
        self.deposit_history_list = []
        self.withdraw_history_list = []

    @staticmethod
    def _generate_account_number():
        # 3자리-2자리-6자리
        first = str(random.randint(100, 999))
        second = str(random.randint(10, 99))
        third = str(random.randint(100000, 999999))
        return f"{first}-{second}-{third}"

    # Q3. 클래스 메서드: 계좌 개수 출력
    @classmethod
    def get_account_num(cls):
        print(f"총 계좌 개수: {cls.account_count}")
        return cls.account_count

    # Q4. 입금 메서드
    def deposit(self, amount):
        if amount < 1:
            print("입금은 최소 1원 이상만 가능합니다.")
            return
        self.balance += amount
        self.deposit_count += 1
        self.deposit_history_list.append(amount)

        # Q7. 5번째 입금 시 이자 지급 (1%)
        if self.deposit_count % 5 == 0:
            interest = int(self.balance * 0.01)
            self.balance += interest
            print(f"입금 5회 달성! 잔고의 1%({interest}원)가 이자로 지급되었습니다.")

    # Q5. 출금 메서드
    def withdraw(self, amount):
        if amount > self.balance:
            print("잔액이 부족합니다.")
            return
        self.balance -= amount
        self.withdraw_history_list.append(amount)

    # Q6. 정보 출력 메서드
    def display_info(self):
        # 잔고에 3자리마다 쉼표
        print(f"은행이름: {self.bank}, 예금주: {self.owner}, 계좌번호: {self.account_number}, 잔고: {self.balance:,}원")

    # Q10. 입금 내역 출력
    def deposit_history(self):
        print("입금 내역:", ', '.join([f"{amt:,}원" for amt in self.deposit_history_list]))

    # Q10. 출금 내역 출력
    def withdraw_history(self):
        print("출금 내역:", ', '.join([f"{amt:,}원" for amt in self.withdraw_history_list]))


# Q8. 여러 객체 생성 및 리스트 저장
acc1 = Account("홍길동", 500_000)
acc2 = Account("김철수", 1_200_000)
acc3 = Account("이영희", 2_000_000)

accounts = [acc1, acc2, acc3]

# 테스트: 입금 & 출금
acc1.deposit(10_000)
acc1.deposit(100_000)
acc1.deposit(50_000)
acc1.deposit(30_000)
acc1.deposit(20_000)  # 5번째 입금 -> 이자 지급

acc2.deposit(200_000)
acc2.withdraw(500_000)

acc3.deposit(1_000_000)
acc3.withdraw(100_000)

# Q9. 잔고가 100만원 이상인 고객 정보 출력
print("\n100만원 이상 잔고 고객:")
for acc in accounts:
    if acc.balance >= 1_000_000:
        acc.display_info()

# Q3. 계좌 개수 출력
Account.get_account_num()

# Q10. 입금/출금 내역 출력
print("\n입금/출금 내역 확인:")
for acc in accounts:
    print(f"\n{acc.owner}님의 내역")
    acc.deposit_history()
    acc.withdraw_history()
```
